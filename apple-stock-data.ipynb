### Stock Shares and Price Fluctuations

A share is the smallest part of a company's stock that can be bought. Share prices fluctuate over time. Using the `history()` method, we can retrieve stock prices over a specific period. The `period` parameter defines the timeframe, with options such as:

- `1d`, `5d` - 1 day, 5 days
- `1mo`, `3mo`, `6mo` - 1, 3, 6 months
- `1y`, `2y`, `5y`, `10y` - 1, 2, 5, 10 years
- `ytd` - Year to date
- `max` - Maximum available data

```python
apple_share_price_data = apple.history(period="max")
apple_share_price_data.head()
```

### Resetting the Index of a DataFrame

The `reset_index()` function in pandas is used to reset the index of a DataFrame. This converts the existing index into a regular column and replaces it with a new default integer-based index.

- **`inplace=True`**: Modifies the original DataFrame directly instead of creating and returning a new one.

#### Example:
Before using `reset_index()`:
| index | A | B |
|-------|---|---|
| x     | 1 | 4 |
| y     | 2 | 5 |
| z     | 3 | 6 |

After using `reset_index(inplace=True)`:
| index | A | B |
|-------|---|---|
| x     | 1 | 4 |
